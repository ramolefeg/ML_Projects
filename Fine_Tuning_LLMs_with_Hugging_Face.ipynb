```python
import torch
from trl import SFTTrainer
from peft import LoraConfig
from datasets import load_dataset
from transformers import (AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig, TrainingArguments, pipeline)
```

## Step 2: Loading the model

```python
llma_model = AutoModelForCausalLM.from_pretrained(pretrained_model_name_or_path ="aboonaji/llama2finetune-v2",quantization_config=BitsAndBytesConfig(load_in_4bit=True,
                                                                                                                                                     bnb_4bit_compute_dtype=getattr(torch,"float16"),
                                                                                                                                                     bnb_4bit_quant_type="nf4"

                                                                                                                                                     ))
llma_model.config.use_cache = False #reduce memory usage and speed up computations
llma_model.config.pretraining_tp =1
```

## Step 3: Loading the tokenizer

```python
llma_tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path ="aboonaji/llama2finetune-v2",trust_remote_code=True)
llma_tokenizer.pad_token = llma_tokenizer.eos_token
llma_tokenizer.padding_side = "right"
```

## Step 4: Setting the training arguments

```python
#defining training hyperparameters
training_arguments = TrainingArguments(output_dir = "./results",per_device_train_batch_size= 4, max_steps=100)
```

## Step 5: Creating the Supervised Fine-Tuning trainer
##Other techniqune RLHF reinforcement learning from human feedback

```python
#from trl import SFTTrainer library
#Training arguments (formated dataset,training_arguments, tokenizer,model)
llma_sft_trainer = SFTTrainer(model =llma_model,
                              args=training_arguments,
                              train_dataset= load_dataset(path ="aboonaji/wiki_medical_terms_llam2_format",split="train"),
                              tokenizer = llma_tokenizer,
                              peft_config =LoraConfig(r= 64,lora_alpha=16,task_type="CAUSAL_LM"),
                              dataset_text_field = "text"
                              )
```

## Step 6: Training the model

```python
#we have all the parameters needed to train the model from the sft_trainer
llma_sft_trainer.train()
```

## Step 7: Chatting with the model

```python
user_prompt = "what is Cachexia"
text_generation_pipeline = pipeline(task ="text-generation", model  =  llma_model,tokenizer=llma_tokenizer,max_length=300)
model_answer = text_generation_pipeline(f"<s>[INST] {user_prompt} [/INST] ")
print(model_answer[0]['generated_text'])
```
